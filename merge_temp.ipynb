```python
ukbb = pd.read_csv("/Users/sonali.gupta/Downloads/Insomnia/Lifted_Hammerschlag_NatGenet2017_insomnia_sumstats-full_090617.txt", sep = "\t")
```

```python
ukbb.head()
```

```python

merged_df = pd.merge(mvp[['SNP_ID','chrom','pos','alt','ref','or','ci']], ukbb[['SNP','POS','A2','A1','BETA','SE']], 
                     left_on=['SNP_ID', 'pos', 'alt', 'ref'], 
                     right_on=['SNP', 'POS', 'A2', 'A1'],
                     how='inner')
```

```python
merged_df
```

```python
# Assuming 'merged_df' is your DataFrame
merged_df['log_OR_x'] = -1*np.log(merged_df['or'])
merged_df[['ci_lower', 'ci_upper']] = merged_df['ci'].str.split(",", expand=True).astype(float)
merged_df['SE_x'] = (np.log(merged_df['ci_upper'])-np.log(merged_df['ci_lower']))/(2*1.96)
#merged_df['BETA'] = -1*merged_df['BETA']
```

```python
merged_df = merged_df[['SNP','chrom','pos','A1','A2','log_OR_x','BETA','SE_x','SE']]
```

```python
merged_df.columns = ['SNP', 'CHR', 'POS', 'REF', 'ALT', 'BETA1', 'BETA2', 'SE1',  'SE2']
```
